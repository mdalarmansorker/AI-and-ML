```python
import os
import zipfile
```

```python
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Define the paths to your folders
folders = {
    'bangla': '/content/drive/My Drive/Colab Notebooks/bangla.zip'  # Adjust path as needed
}

# Check if the file exists
if os.path.exists(folders['bangla']):
    print("File found!")
else:
    print("File not found.")
```

```python
# Function to load data and labels
def load_data(folders):
    texts = []
    labels = []

    # Loop over each folder and read the files
    for label, folder in folders.items():
        # Check if it's a zip file
        if folder.endswith('.zip'):
            with zipfile.ZipFile(folder, 'r') as zip_ref:
                for filename in zip_ref.namelist():
                    with zip_ref.open(filename, 'r') as file:
                        text = file.read().decode('utf-8')  # Decode if necessary
                    texts.append(text)
                    labels.append(label)
        else:
            # If it's a directory, handle it as before
            for filename in os.listdir(folder):
                file_path = os.path.join(folder, filename)
                with open(file_path, 'r', encoding='utf-8') as file:
                    text = file.read()
                texts.append(text)
                labels.append(label)

    return texts, labels
```

```python
# Load the data
texts, labels = load_data(folders)
texts
labels
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
```

```python
# Transform the text data into TF-IDF features
X = vectorizer.fit_transform(texts)

# Check the shape of the resulting matrix
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

# Check the shape of X before splitting (this should give you (number_of_samples, number_of_features))
print(X.shape)
```

```python
# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.4, random_state=42)
# Check the size of the splits
print(f"Training samples: {X_train.shape[0]}, Testing samples: {X_test.shape[0]}")
```

```python
from sklearn.tree import DecisionTreeClassifier

# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the model
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Make predictions on the test set
y_pred = clf.predict(X_test)
y_pred
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
```

```python
# Get the classification report (precision, recall, F1-score)
print("Classification Report:")
print(classification_report(y_test, y_pred))# Get the classification report (precision, recall, F1-score)
print("Classification Report:")
print(classification_report(y_test, y_pred))
```
